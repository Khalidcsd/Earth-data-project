## Set up an anaconda installation

### Download Anaconda3

Get the latest Anaconda3 installer from https://www.anaconda.com/download/

### Install Miniconda3

following instructions for relevant platforms

### Install the environment

There is an environment file [introearth.yml](piggy/Spring2019/DE/DE-Earth/Extra/)

containing (hopefully) everything needed to run lab materials:


please download locally, e.g. in your ```$HOME``` directory the file ```introearth.yml```

With conda installed, create the environment using

```bash
conda env create --file introearth.yml
```

### Activate the environment

After creating the environment, activate it using:

```bash
source activate introearth
```

### De-activate the environment

To de-activate the environment, use:

```
source deactivate 
```
